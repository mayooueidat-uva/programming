```python
# covariance
import itertools # so i can iterate over multiple lists at once.

def y_bar(y):
  sample_mean = (np.sum(y))/(len(y))
  return sample_mean
y_bar(y)

def cov(x, y):
  x_vals = []
  y_vals = []
  xy_vals = []
  for i in x:
    # calculating each (xn-xbar) value for numerator
    x_val = i-x_bar(x)
    # appending to a list
    x_vals.append(x_val)
  for i in y:
    # calculating each (yn-ybar) value for numerator
    y_val = i-y_bar(y)
    # appending to a list
    y_vals.append(y_val)
  for (i, n) in zip(x_vals, y_vals):
    # calculating each (xn-xbar)*(yn-ybar) value in numerator
    xy_val = i*n
    xy_vals.append(xy_val)
  # summing up all values in numerator, dividing them by N-1
  return (np.sum(xy_vals)/(len(xy_vals)-1))
cov(x,y)
```

```python
# stdev of y
def s_stdev_y(y):
    vals_for_sd_numerator = []
    for i in y:
      val = (i-(y_bar(y)))**(2)
      vals_for_sd_numerator.append(val)
    numerator_1 = np.sum(vals_for_sd_numerator)
    denominator_1 = (len(y)-1)
    sd = (numerator_1/denominator_1)**(1/2)
    return sd
s_stdev_y(y)

# sample correlation coefficient
# (this is what is actually getting printed)
def r(x,y):
  denominator = s_stdev_x(x)*s_stdev_y(y)
  numerator = cov(x,y)
  r = numerator/denominator
  return r
r(x,y)
```
